#Hartree-Fock code for closed-shell systems

Consider a Slater determinant built up of single-particle orbitals $\psi_{\lambda}$, 
with $\lambda = 1,2,\dots,N$, where $N$ is number of particles or equivalently the number of spin-orbitals. The expand each spin-orbital occurding to

$$
\psi_i  = \sum_{\lambda} C_{i\lambda}\phi_{\lambda},
$$

##The Fock operator
$$
h_{\alpha\beta}^{HF}=\langle \alpha | h | \beta \rangle+
\sum_{i=1}^N\sum_{\gamma\delta} C^*_{i\gamma}C_{i\delta}\langle \alpha\gamma|V|\beta\delta\rangle_{AS}
$$



    for(int alpha = 0; alpha < nStates; alpha++){

                    }
                }
            }
            H(alpha, beta) = onebodyElement(Z, alpha, beta) + interaction;
        }
    
    }

##One-electron operator
$$
\langle \alpha | h | \beta \rangle = \int \phi_{\alpha}^*(\mathbf{x}) h \phi_{\beta}(\mathbf{x})\; d\mathbf{x} =
\left\{ 
  \begin{array}{l l}
    -\frac{Z^2}{2n^2} & \quad \text{if $\alpha = \beta$}\\
    0 & \quad \text{if $\alpha \neq \beta$}
  \end{array} \right.
$$


    double onebodyElement(const int Z, const int alpha, const int beta)
    {
        double result;
    
        if(alpha==beta){
            int n = mapStateToPrincipleQuantumNumber(alpha);
            result = -Z*Z/(2.0 * n * n);
        }else{
            result = 0.0;
        }
    
        return result;
    
    }
---

##Two-electron operator
$$
\langle \alpha\gamma|V|\beta\delta\rangle_{AS} 
= \int \phi_{\alpha}^*(\mathbf{x}_i)\phi_{\gamma}^*(\mathbf{x}_j) V \phi_{\beta}(\mathbf{x}_i)\phi_{\delta}(\mathbf{x}_j)\; d\mathbf{x}_id\mathbf{x}_j - \int \phi_{\alpha}^*(\mathbf{x}_i)\phi_{\gamma}^*(\mathbf{x}_j) V \phi_{\delta}(\mathbf{x}_i)\phi_{\beta}(\mathbf{x}_j)\; d\mathbf{x}_id\mathbf{x}_j
$$

$$
Q_{\alpha\gamma\beta\delta} 
= \int \varphi_{\alpha}^*(\mathbf{r}_i)\varphi_{\gamma}^*(\mathbf{r}_j) V \varphi_{\beta}(\mathbf{r}_i)\varphi_{\delta}(\mathbf{r}_j)\; d\mathbf{r}_id\mathbf{r}_j - \int \varphi_{\alpha}^*(\mathbf{r}_i)\varphi_{\gamma}^*(\mathbf{r}_j) V \varphi_{\delta}(\mathbf{r}_i)\varphi_{\beta}(\mathbf{r}_j)\; d\mathbf{r}_id\mathbf{r}_j
$$


    double twobodyElement(const field<mat> Q, int alpha, int beta, int gamma, int delta)
    {
    
        double directTerm = 0;
        double exchangeTerm = 0;
        int n1 = mapStateToPrincipleQuantumNumber(alpha) - 1;
        int n2 = mapStateToPrincipleQuantumNumber(beta)  - 1;
        int n3 = mapStateToPrincipleQuantumNumber(gamma) - 1;
        int n4 = mapStateToPrincipleQuantumNumber(delta) - 1;
    
        if(pow(-1,alpha) == pow(-1,beta) && pow(-1,gamma) == pow(-1,delta)) {
            directTerm  += Q(n1, n3)(n2, n4);
        }
        if(pow(-1,alpha) == pow(-1,delta) && pow(-1,beta) == pow(-1,gamma)) {
            exchangeTerm -= Q(n1, n3)(n4, n2);
        }
    
        return directTerm + exchangeTerm ;
    }

###Energy

$$
E^{HF}=\sum_{i=1}^N\sum_{\alpha\beta}C^*_{i\alpha}C_{i\beta}\langle \alpha | h | \beta \rangle
+\frac{1}{2}\sum_{ij=1}^N\sum_{\alpha\beta\gamma\delta} C^*_{i\alpha}C^*_{j\gamma}C_{i\beta}C_{j\delta}\langle \alpha\gamma|V|\beta\delta\rangle_{AS}
$$

    //The one-body term:
    for (int i = 0; i < nParticles; i++){
        for (int alpha = 0; alpha < nStates; alpha++){
            E0 += C(i,alpha)*C(i,alpha)*onebodyElement(Z, alpha, alpha);
        }
    }

    //The two-body term:
    for (int i = 0; i < nParticles; i++){
        for(int j = 0; j < nParticles; j++){

            for (int alpha = 0; alpha < nStates; alpha++){
                for (int beta = 0; beta < nStates; beta++){
                    for (int gamma = 0; gamma < nStates; gamma++){
                        for (int delta = 0; delta < nStates; delta++){

                            double ECoulumb=twobodyElement(Q,alpha, beta, delta, gamma);
                            E0 += 0.5*C(i,alpha)*C(j,gamma)*C(i,beta)*C(j,delta)*ECoulumb;
                        }
                    }

                }
            }
        }
    }

    int mapStateToPrincipleQuantumNumber(const int state)
    {
        vec map;
        map << 1 << 1 << 2 << 2 << 3 << 3 << 4 << 4;
    
        int n = map(state);
        return n;
    }
